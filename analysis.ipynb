# Project Management Analysis

This notebook provides an exploratory data analysis (EDA) and predictive modeling for a synthetic dataset representing project management scenarios. The dataset includes fields such as project start and end dates, budget, team size, complexity level, manager experience, client satisfaction, quality score, whether the project was delayed, and if it was successful. The goal is to explore the data visually and statistically and build a predictive model for project success.



```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load dataset
df = pd.read_csv('synthetic_project_data.csv', parse_dates=['StartDate','EndDate'])

# Display the first few rows
df.head()
```

```
# Summary statistics for numeric columns
df.describe(include=[np.number])
```

```
# Histogram of budgets
plt.figure(figsize=(8,4))
sns.histplot(df['Budget'], kde=True)
plt.title('Budget Distribution')
plt.xlabel('Budget')
plt.ylabel('Frequency')
plt.show()

# Bar plot for project success counts
plt.figure(figsize=(5,4))
sns.countplot(x='Success', data=df)
plt.title('Project Success Counts')
plt.xlabel('Success (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()

# Box plot of Quality Score by Success
plt.figure(figsize=(8,4))
sns.boxplot(x='Success', y='QualityScore', data=df)
plt.title('Quality Score by Project Success')
plt.xlabel('Success (0=No, 1=Yes)')
plt.ylabel('Quality Score')
plt.show()
```

```
# Correlation matrix
numeric_cols = ['DurationDays','Budget','TeamSize','ComplexityLevel','ManagerExperienceYears','ClientSatisfaction','QualityScore','Delayed','Success']

corr = df[numeric_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```
# Prepare features and target
X = df[['DurationDays','Budget','TeamSize','ComplexityLevel','ManagerExperienceYears','ClientSatisfaction','QualityScore','Delayed']]
y = df['Success']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Logistic Regression Model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_scaled, y_train)

# Predictions
y_pred = log_reg.predict(X_test_scaled)

# Evaluation
print('Accuracy:', accuracy_score(y_test, y_pred))
print('
Classification Report:
', classification_report(y_test, y_pred))
print('
Confusion Matrix:
', confusion_matrix(y_test, y_pred))
```
